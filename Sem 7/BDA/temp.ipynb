```python
import pandas as pd
import rpy2.robjects as robjects
from rpy2.robjects import pandas2ri
from rpy2.robjects.conversion import localconverter

# Step 1: Load dataset in Python
df = pd.read_csv("Tweets.csv")

# Preview dataset
print("Sample data:")
print(df.head())

# Step 2: Transfer DataFrame to R using localconverter
with localconverter(robjects.default_converter + pandas2ri.converter):
    robjects.globalenv['tweets_df'] = robjects.conversion.py2rpy(df)

# Step 3: Define R code block for SMA
r_code = """
library(ggplot2)
library(syuzhet)
library(wordcloud)
library(dplyr)
library(RColorBrewer)

tweets_text <- tweets_df$text

# Sentiment Analysis
sentiments <- get_nrc_sentiment(tweets_text)
sentiment_scores <- colSums(sentiments)

# Sentiment Bar Plot
barplot(sentiment_scores,
        col=rainbow(10),
        las=2,
        main="Sentiment Distribution for Tweets Dataset")

# Positive vs Negative Pie Chart
pos_neg <- c(
  Positive = sum(sentiments$positive),
  Negative = sum(sentiments$negative)
)
pie(pos_neg, labels = names(pos_neg), col = c("green", "red"),
    main = "Positive vs Negative Tweets")

# Word Cloud
words <- tolower(tweets_text)
wordcloud(words, max.words=100, random.order=FALSE, colors=brewer.pal(8, "Dark2"))
"""

# Step 4: Execute R code from Python
robjects.r(r_code)
```
