```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, color="blue", linestyle="-", marker="o", label="sin(x)")
plt.plot(x, y2, color="red", linestyle="--", marker="s", label="cos(x)")

plt.title("Styled Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.grid(True, linestyle=":", alpha=0.7)
plt.savefig("line_styles.png", dpi=200)
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, s=100, c="orange", marker="^", edgecolor="black", alpha=0.8)
plt.title("Styled Scatter Plot")
plt.savefig("scatter_styles.png", dpi=200)
plt.show()
```

```python
categories = ["A", "B", "C", "D"]
values = [4, 7, 3, 8]
colors = ["skyblue", "salmon", "limegreen", "violet"]

plt.bar(categories, values, color=colors, edgecolor="black")
plt.title("Styled Bar Chart")
plt.savefig("bar_styles.png", dpi=200)
plt.show()
```

```python
data = np.random.randn(1000)

plt.hist(data, bins=30, color="teal", alpha=0.6, edgecolor="black")
plt.title("Styled Histogram")
plt.savefig("hist_styles.png", dpi=200)
plt.show()
```

```python
x = np.linspace(0, 10, 100)

# Apply a global style
plt.style.use("seaborn-v0_8")

fig, axs = plt.subplots(2, 2, figsize=(8, 6))

axs[0,0].plot(x, np.sin(x), "b--o"); axs[0,0].set_title("Sine")
axs[0,1].plot(x, np.cos(x), "r-.*"); axs[0,1].set_title("Cosine")
axs[1,0].scatter(np.random.rand(50), np.random.rand(50), c="green", marker="x"); axs[1,0].set_title("Scatter")
axs[1,1].hist(np.random.randn(100), bins=20, color="purple", alpha=0.7); axs[1,1].set_title("Histogram")

plt.tight_layout()
plt.savefig("subplots_styles.png", dpi=200)
plt.show()
```
